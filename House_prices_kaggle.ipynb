```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

train = pd.read_csv("/Users/TylerYoung/Desktop/Desktop/LEARNPYTHON/House Prices/train.csv")

test = pd.read_csv("/Users/TylerYoung/Desktop/Desktop/LEARNPYTHON/House Prices/test.csv")
```

```python
print(train)
print(test)
```

```python
train_describe = train.describe()
train_shape = train.shape

test_describe = test.describe()
test_shape = test.shape

print(train_describe)
print(train_shape)
print(test_describe)
print(test_shape)
```

```python
#Filling NA's in Lot Frontage to Zero
train["LotFrontage"] = train["LotFrontage"].fillna(0)
test["LotFrontage"] = test["LotFrontage"].fillna(0)
```

```python
#Take Forest features wanted
target = train["SalePrice"]
features_forest = train[["Id", "MSSubClass", "LotFrontage", "LotArea", "OverallQual", "YearBuilt", "YearRemodAdd", "WoodDeckSF", 
                "OpenPorchSF","3SsnPorch", "ScreenPorch", "PoolArea", "MiscVal", "MoSold" , "YrSold"]]

#Building and fitting forest
forest = RandomForestClassifier(max_depth = 10, min_samples_split = 2, n_estimators = 100, random_state = 1)
my_forest = forest.fit(features_forest,target)

print(my_forest.score(features_forest, target))

#Compute predictions on test set features then print the length of the prediction vector
test_features = test[["Id", "MSSubClass", "LotFrontage", "LotArea", "OverallQual", "YearBuilt", "YearRemodAdd", "WoodDeckSF", 
                "OpenPorchSF","3SsnPorch", "ScreenPorch", "PoolArea", "MiscVal", "MoSold" , "YrSold"]]
forest_prediction = my_forest.predict(test_features)
print(len(forest_prediciton))
```

```python
#Printing random forrest feature and score
print(my_forest.feature_importances_)
print(my_forest.score(features_forest, target))
```

```python
import numpy as np

# Create a data frame with two columns: Id & forest_prediction
Id = np.array(test["Id"]).astype(int)
House_solution_forest = pd.DataFrame(forest_prediction, Id, columns = ["SalePrice"])

# Check that your data frame has 418 entries
print(House_solution_forest)

#create the solution array and create .csv file
House_solution_forest.to_csv("House_solution_forest.csv", index_label = ["Id"])
```
